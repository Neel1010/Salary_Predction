```python
df = df[df["Salary"].notnull()]
df.head()
```

```python
df.info()
```

```python
df=df[df['Employment']=='Employed full-time']
```

```python
df['EdLevel'].value_counts()
```

```python
df=df.dropna()
```

```python
def clean_Ed(x):
    if 'Bachelor’s degree' in x:
        return 'Bachelor’s degree'
    if 'Master’s degree' in x:
        return 'Master’s degree'
    if 'Professional degree' in x or 'Other doctoral' in x:
        return 'Post grad'
    return 'Less than a Bachelors'
```

```python
df['EdLevel']=df['EdLevel'].apply(clean_Ed)
df
```

```python
def shorten_categories(categories, cutoff):
    categorical_map = {}
    for i in range(len(categories)):
        if categories.values[i] >= cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            categorical_map[categories.index[i]] = 'Other'
    return categorical_map
```

```python
country_map = shorten_categories(df.Country.value_counts(), 400)
df['Country'] = df['Country'].map(country_map)
df.Country.value_counts()
```

```python
df
```

```python
df['YearsCodePro'].unique()
```

```python
def clean_exp(x):
    if x=='Less than 1 year':
        return 0.5
    if x=='More than 50 years':
        return 50
    return x
```

```python
df['YearsCodePro'] = df['YearsCodePro'].apply(clean_exp)
```

```python
df=df[df['Salary']<=250000]
df=df[df['Salary']>=10000]
```

```python
df
```

```python
df.drop(['Employment'],axis=1,inplace=True)
df
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le_education = LabelEncoder()
df['EdLevel'] = le_education.fit_transform(df['EdLevel'])
df["EdLevel"].unique()
```

```python
le_country = LabelEncoder()
df['Country'] = le_country.fit_transform(df['Country'])
df["Country"].unique()
```

```python
df
```

```python
X=df[['Country','EdLevel','YearsCodePro']]
y=df['Salary']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
```

```python
predictions=linear_reg.predict(X_test)
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

error = np.sqrt(mean_squared_error(y_test, predictions))
```

```python
error
```

```python
from sklearn.tree import DecisionTreeRegressor
dec_tree_reg = DecisionTreeRegressor(random_state=10)
dec_tree_reg.fit(X_train, y_train)
```

```python
y_pred = dec_tree_reg.predict(X)
```

```python
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest_reg = RandomForestRegressor(random_state=0,n_estimators=100)
random_forest_reg.fit(X, y.values)
y_pred = random_forest_reg.predict(X)
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```
